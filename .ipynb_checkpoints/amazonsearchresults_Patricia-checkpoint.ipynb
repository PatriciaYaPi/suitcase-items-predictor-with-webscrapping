```python
from selectorlib import Extractor
import requests 
import json 
from time import sleep
import pandas as pd
```

```python
# INPUTS:
#########

#LIST OF ITEMS TO SEARCH:
#words_to_search=['summer+dress+women', 'shorts+women', 't-shirt+women', 'mosquito+repellent+spray']

#open and read the file with the items listed for my possible suitcase:
#file = open("/mysuitcase.txt", "r")
#words_to_search=file.read()
#words_to_search

my_file = open("suitcase_items_to_search_on_amazon.txt", "r") 
content = my_file.read()
words_to_search = content.split(",")
my_file.close()


"""
Those words will be search on:
------------------------------
https://www.amazon.com/s?k=shorts
https://www.amazon.com/s?k=hat
https://www.amazon.com/s?k=sunglasses
"""

#NUMBER OF ITEMS OF EACH CLASS TO SEARCH:
limit = 3  #Idea then use one random item of the list. So a different suitcase each time you "make" the suitcase

store_web = 'https://www.amazon.com/s?k='

#yml file for web scrapping: search_results.yml

# OUTPUT
########
# JSON FILE WITH A LIST OF DICTIONARIES: search_results_output.jsonl
# with product info: title, price, amazon rating, img, url
```

```python
# Create an Extractor by reading from the YAML file
e = Extractor.from_yaml_file('search_results.yml')
```

```python
def scrape(url):  

    headers = {
        'dnt': '1',
        'upgrade-insecure-requests': '1',
        'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.61 Safari/537.36',
        'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
        'sec-fetch-site': 'same-origin',
        'sec-fetch-mode': 'navigate',
        'sec-fetch-user': '?1',
        'sec-fetch-dest': 'document',
        'referer': 'https://www.amazon.com/',
        'accept-language': 'en-GB,en-US;q=0.9,en;q=0.8',
    }

    # Download the page using requests
    print("Downloading %s"%url)
    r = requests.get(url, headers=headers)
    # Simple check to check if page was blocked (Usually 503)
    if r.status_code > 500:
        if "To discuss automated access to Amazon data please contact" in r.text:
            print("Page %s was blocked by Amazon. Please try using better proxies\n"%url)
        else:
            print("Page %s must have been blocked by Amazon as the status code was %d"%(url,r.status_code))
        return None
    # Pass the HTML of the page and create 
    return e.extract(r.text)
```

```python
product_data = []
#with open("search_results_urls.txt",'r') as urllist, open('search_results_output.jsonl','w') as outfile:
with open('search_results_output.jsonl','w') as outfile:
    #for url in urllist.read().splitlines():
    for item in words_to_search:
        url = store_web+item
        data = scrape(url) 
        if data:
            for index,product in enumerate(data['products']):
                #Just take  a number "limit" items:
                if index == limit:
                    break
                product['search_url'] = url
                print("Saving Product: %s"%product['title'])
                product['item'] = item
                json.dump(product,outfile) #write on json file
                outfile.write("\n")
                
                product_data.append(product)
                sleep(3)
                
```

```python
# ITEMS FOUND IN PANDA FORMAT:
##############################
search_results_output_pd = pd.DataFrame(product_data)
search_results_output_pd
```

```python
#Get a sample of each product type:
#So we get a new suitcase each time

random_suitcase=search_results_output_pd.groupby('item').apply(lambda x: x.sample(1)).reset_index(drop=True)
```

```python
#Output random_suitcase into a jsonl file:

dictionary_varibale = [ 
    dict([
        (colname, row[i]) 
        for i,colname in enumerate(random_suitcase.columns)
    ])
    for row in random_suitcase.values
]
print(json.dumps(dictionary_varibale))

with open('suitcase_items_randomly_from_amazon.jsonl','w') as outfile:
    json.dump(dictionary_varibale,outfile) #write on json file
```
